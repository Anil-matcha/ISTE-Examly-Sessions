```python
import numpy as np
import pandas as pd
import seaborn as sns 
```

```python
cars = pd.read_csv("cars93.csv")
cars.head()
```

### Box Plot

```python
sns.boxplot(x="Price", data=cars,palette='rainbow', orient = "v")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x="Price", data=cars,palette='rainbow', orient = "v", width = 0.3)
```

```python
sns.boxplot(x=cars["Price"],palette='rainbow', orient = "v")
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(data=cars,palette='rainbow')
```

```python
cars.columns
```

```python
sns.boxplot(x="AirBags", y="Price",data=cars, palette="rainbow")
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Type", y="Price",hue = "AirBags", data=cars, palette="rainbow")
```

### Histogram

```python
sns.distplot(cars["Horsepower"])
```

```python
sns.distplot(cars["Horsepower"],kde=False)
```

```python
sns.distplot(cars["Horsepower"],kde=False, bins = 16)
```

#### Symmetric and Skewed data

```python
a = np.random.randn(100000)
sns.distplot(a,hist = False,kde=True)
```

```python
sns.distplot(cars["Price"],hist = False,kde=True)
```

```python
sns.distplot(cars["Horsepower"],hist = False,kde=True)
```

```python
from scipy.stats import skew
print("Skewness of height parameter : ", skew(cars["Price"]))
print("Skewness of height parameter : ", skew(cars["Horsepower"]))
```

### Bar Chart

This is essentially the same as barplot

```python
sns.countplot(cars["Type"])
```

```python
sns.barplot(x='Type',y='MPG.city',data=cars)
```

```python
sns.barplot(x='Type',y='MPG.city',data=cars,estimator=np.std)
```

### Scatter plot

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Horsepower',y='MPG.city',data = cars, hue = "AirBags", palette = "Set1")
```

### Jointplot

jointplot() allows to basically match up(joint) two distplots for bivariate analysis

```python
sns.jointplot(x='Horsepower',y='MPG.city',data=cars,kind='scatter')
```

```python
sns.jointplot(x='Horsepower',y='MPG.city',data=cars,kind='reg')
```

### Pairplot

To plot pairwise relationships across the entire dataframe(only for numeric columns) 

```python
sns.pairplot(cars)
```
