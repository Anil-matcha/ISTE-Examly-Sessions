Refer to the **tips.csv** dataset. The dataset contains total 244 tips(data about customers served by a restaurant) across 7 different columns.

* **totbill** - Total bill (cost of the meal), including tax, in US dollars
* **tip** - Tip (gratuity) in US dollars
* **sex** - Sex of person paying for the meal (0=male, 1=female)
* **smoker** - Smoker in party? (0=No, 1=Yes)
* **day** - Sun, Sat, Thu, Fri
* **time** - Night and Day
* **size** - Size of the party

#### Q1. Plot a day wise boxplot of totbill column as shown below. 

Note - In all the visualizations here, the rainbow palette is used.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv("tips.csv")
tips.head()
```

```python
sns.boxplot(x = "day", y = "totbill", data = tips, palette = "rainbow")
plt.title("Day wise Boxplot of totbill")
plt.show()
```

#### Q2. Plot a pariplot by setting hue parameter to sex.

```python
sns.pairplot(data = tips, hue = "sex",palette = "rainbow")
plt.show()
```

#### Q3. Draw a scatter plot of tip and totalbill and also allot the points according to the sex.

```python
plt.figure(figsize = (12,8))
sns.scatterplot(x = "tip", y = "totbill", hue = "sex", data = tips,palette = "Set1")
plt.title("Scatterplot of tip vs. totbill")
plt.show()
```

#### Q4. Plot a hitsogram of totbill for 50 bins. 

```python
sns.distplot(tips["totbill"],bins = 50, kde = False)
plt.title("Histogram of totbill")
plt.show()
```

#### Q5. Plot a heatmap to represent entire dataset's correlation.

```python
sns.heatmap(tips.corr(), cmap='rainbow', annot = True)
plt.title("tips data correlation")
plt.show()
```
